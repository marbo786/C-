```python
# data
import numpy as np
X = np.linspace(1,20)
Y = np.exp(X)

# line plot
plt.plot(X,Y)

# display the plot
```

From the plot, it can be observed that as 'X' is increasing there is an exponential increase in Y.

**The above plot can be represented not only by a solid line, but also a dotted line with varied thickness. The points can be marked explicitly using any symbol.**

```python
# data
X = np.linspace(1,20,100)
Y = np.exp(X)

# line plot
# the argument 'r*' plots each point as a red '*' 
plt.plot(X,Y, '*')

# display the plot
plt.show()
```

We can change the colors or shapes of the data points.

There can be multiple line plots in one plot. Let's plot three plots together in a single graph. Also, add a plot title.

```python
# data
X = np.linspace(1,20)
Y1 = X
Y2 = np.square(X)
Y3 = np.sqrt(X)

# line plot
plt.plot(X,Y1,'r', X,Y2,'b', X,Y3,'y')

# add title to the plot
plt.title('Line Plot')

# display the plot
plt.show()
```

### Line chart application Example for students

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample time series data (dates and closing prices)
dates = pd.date_range(start="2023-01-01", periods=30, freq="D")  # Daily data for 30 days
closing_prices = [150 + i*2 + (i % 5)*3 for i in range(30)]  # Simulated stock prices

# Create a line plot
plt.figure(figsize=(10, 5))
plt.plot(dates, closing_prices, color='blue', marker='o', linestyle='-', linewidth=2, markersize=5)

# Customize the plot
plt.title("Apple Stock Closing Prices (Jan 2023)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price ($)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.1)
plt.xticks(rotation=45)  # Rotate dates for readability

# Highlight a specific event (e.g., peak price)
peak_price = max(closing_prices)
peak_date = dates[closing_prices.index(peak_price)]
plt.annotate(f'Peak: ${peak_price}', 
             xy=(peak_date, peak_price), 
             xytext=(peak_date + pd.Timedelta(days=2), peak_price - 5),
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.tight_layout()  # Prevent overlapping elements
plt.show()
```

```python
tips_data = pd.read_csv("tips.csv")
```

```python
# check the head() of the tips dataset
tips_data.head()
```

Plot the scatter plot for the variables 'total_bill' and 'tip'.

```python
# data
X = tips_data['total_bill']
Y = tips_data['tip']

# plot the scatter plot
plt.scatter(X,Y)

# add the axes labels to the plot
plt.xlabel('total_bill')
plt.ylabel('tip')

# display the plot
plt.show()
```

We can add different colors, opacity, and shapes to data points. Let's add these customizations to the above plot.

```python
# plot the scatter plot for the variables 'total_bill' and 'tip'

X = tips_data['total_bill']
Y = tips_data['tip']

# plot the scatter plot
# s is for size of marker, c is for colour, alpha is for opacity (0 < alpha < 1)
plt.scatter(X, Y, s = np.array(Y)**3, c= 'green',marker = 's', alpha= 0.5)
 
# add title 
plt.title('Scatter Plot')

# add the axes labels to the plot
plt.xlabel('total_bill')
plt.ylabel('tip')

# display the plot
plt.show()
```

The bubbles with greater radius display that the tip amount is more as compared to the bubbles with less radius.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots (1 row, 2 columns)
plt.subplot(1, 2, 1)  # 1st subplot
plt.plot(x, y1, c='r')
plt.title("Sine")

plt.subplot(1, 2, 2)  # 2nd subplot
plt.plot(x, y2, c='y')
plt.title("Cosine")

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Plot in each subplot
axs[0].plot(x, y1,)
axs[0].set_title("Sine")

axs[1].plot(x, y2)
axs[1].set_title("Cosine")


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Create figure with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

# Plot in each subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title("Sine")

axs[0, 1].plot(x, y2)
axs[0, 1].set_title("Cosine")

axs[1, 0].plot(x, y3)
axs[1, 0].set_title("Tangent")
axs[1, 0].set_ylim(-10, 10)  # optional to limit tangent spikes

axs[1, 1].plot(x, y4)
axs[1, 1].set_title("Exponential Decay")

# Adjust spacing
plt.tight_layout()
plt.show()
```

### 2.4 Pie Plot
<table align="left">
    <tr>
        <td>
            <div align="left", style="font-size:120%">
                <font color="#21618C">
                    <b>Pie plot is a graphical representation of univariate data. It is a circular graph divided into slices displaying the numerical proportion. For the categorical variable, each slice of the pie plot corresponds to each of the categories.
</b>
                </font>
            </div>
        </td>
    </tr>
</table>



```python
# check the head() of the tips dataset
tips_data.head()
```

```python
# categories in the 'day' variable
tips_data.day.value_counts()
```

```python
# plot the occurrence of different days in the dataset

# 'autopct' displays the percentage upto 1 decimal place
# 'radius' sets the radius of the pie plot
plt.pie(tips_data.day.value_counts(), autopct = '%.1f%%', radius = 1.2, labels = ['Sat', 'Sun','Thur','Fri'])

# display the plot
plt.show()
```

From the above pie plot, it can be seen that the data has a high proportion for Saturday followed by Sunday.

**Exploded pie plot** is a plot in which one or more sectors are separated from the disc.

```python
# plot the occurrence of different days in the dataset

# exploded pie plot
plt.pie(tips_data.day.value_counts(), autopct = '%.1f%%', radius = 1.2, labels = ['Sat', 'Sun','Thur','Fri'],
        explode = [0,0.6,0,0.6])

# display the plot
plt.show()
```

**Donut pie plot** is a type of pie plot in which there is a hollow center representing a doughnut.

```python
# plot the occurrence of different days in the dataset

# pie plot
plt.pie(tips_data.day.value_counts(), autopct = '%.1f%%', radius = 1.2, labels = ['Sat', 'Sun','Thur','Fri'])

# add a circle at the center
circle =  plt.Circle( (0,0), 0.5, color='white')
plot = plt.gcf()
plot.gca().add_artist(circle)
 
# display the plot
plt.show()
```
